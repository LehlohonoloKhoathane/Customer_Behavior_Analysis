# Exploratory Data Analysis

```python
# Import the pandas library for data manipulation and analysis
import pandas as pd
# Load the CSV file into a pandas DataFrame. The dataset 'customer_shopping_behavior.csv' is read from the current working directory
df = pd.read_csv('customer_shopping_behavior.csv') #reading the file
```

```python
# Display the first 5 rows of the DataFrame, Used to quickly inspect the structure, column values, and sample records.
df.head()
```

```python
# Display a concise summary of the DataFrame structure, including column names, data types, non-null counts, and memory usage.
df.info()
```

```python
# Generate descriptive statistics for all columns in the DataFrame,including both numerical and categorical features.
df.describe(include='all')
```

```python
# Check for missing values in each column of the DataFrame. Returns the total number of null (NaN) values per column.
df.isnull().sum()
```

```python
# For each 'Category', fill missing values (NaN) in 'Review Rating'  with the median 'Review Rating' of that specific category.
# Uses groupby to segment the data and transform to preserve original DataFrame shape.
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Check for missing values in each column of the DataFrame. Returns the total number of null (NaN) values per column.
df.isnull().sum() 
```

```python
# Convert all column names in the DataFrame to lowercase. This standardizes column naming and helps avoid case-sensitivity issues
df.columns = df.columns.str.lower()
```

```python
# Replace spaces in column names with underscores. Improving consistency and making column names easier to reference in code
df.columns = df.columns.str.replace(' ','_')
```

```python
# Rename the column 'purchase_amount_(usd)' to a cleaner and more readable name 'purchase_amount'
# Improves consistency and simplifies column referencing in analysis
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
# Retrieve and display the column names of the DataFrame, Returns a pandas Index object containing all column labels.
df.columns
```

```python
# Define custom labels for age groups that will be used when binning ages
# These labels correspond to quantile-based or interval-based age categories
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
```

```python
# Dividing the 'age' column into 4 equal-sized groups (quartiles)and assign each group a label from the 'labels' list. 
# Using quantile-based binning so each group has roughly the same number of records.
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
# Displaying the first 10 rows of the 'age' and 'age_group' columns to inspect how ages are grouped into categories
df[['age', 'age_group']].head(10)
```

```python
# Create a mapping dictionary to convert payment frequency labels into their equivalent number of days.
frequency_mapping = {
    'Fortnightly': 14,
    'weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
```

```python
# Map the 'frequency_of_purchases' column to numeric values (in days) using the predefined frequency_mapping dictionary. 
# Converts categorical frequency labels into a numerical feature. Any values not found in the dictionary will be set to NaN.
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# Display the first 10 rows of both the original frequency columnand the newly created numeric version (in days) to
#verify the mapping worked correctly.
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# Display the first 10 rows of 'discount_applied' and 'promo_code_used' to inspect their values and verify data consistency between discount usage
# and promotional code usage.
df[['discount_applied','promo_code_used']].head(10)
```

```python
# Check if all values in 'discount_applied' and 'promo_code_used' are identical row by row. Returns True if every row matches, otherwise False.
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Remove the 'promo_code_used' column from the DataFrame, axis=1 specifies that a column (not a row) is being dropped.
# The result is reassigned back to df to persist the change.
df = df.drop('promo_code_used', axis=1)
```

```python
# Import create_engine function from SQLAlchemy to establish a connection to a database
from sqlalchemy import create_engine
```

```python
# Create a SQLAlchemy engine to connect to a local SQL Server Express instance using Windows Authentication (trusted_connection=yes)
engine = create_engine(
    "mssql+pyodbc://localhost\\SQLEXPRESS/customer_behavior?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"
)

# Write the DataFrame (df) to SQL Server as a table called 'customer_data', if_exists='replace' means it will overwrite the table if it already exists
# index=False prevents pandas from writing the DataFrame index as a column
df.to_sql('customer_data', engine, if_exists='replace', index=False)
```
